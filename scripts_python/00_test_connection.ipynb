```python
# 2. Hứng các giá trị cấu hình
host = os.getenv('MYSQL_SERVER')
database = os.getenv('MYSQL_DB')
username = os.getenv('MYSQL_USER')
password = os.getenv('MYSQL_PASS')
```

```python
# 3. Mở kết nối đến MySQL và kiểm tra
try:
    print(f"Đang tiến hành kết nối vào Server: {host}...")
    
    conn = mysql.connector.connect(
        host=host,
        database=database,
        user=username,
        password=password
    )
    
    if conn.is_connected():
        print(f"✅ Kết nối MySQL thành công! Đã truy cập vào database: {database}")
        
        # Luôn nhớ đóng kết nối sau khi làm việc xong để giải phóng bộ nhớ
        conn.close()
        
except mysql.connector.Error as err:
    print(f"❌ Kết nối thất bại. Lỗi chi tiết: {err}")
```

```python
import pandas as pd
import warnings
```

```python
# Tắt cảnh báo UserWarning của Pandas khi dùng trực tiếp mysql.connector
warnings.filterwarnings('ignore', category=UserWarning)

```

```python
# 1. Viết câu lệnh SQL (Ví dụ: lấy thử 5 dòng đầu tiên để xem trước)
query = "SELECT * FROM SaaS_Catalog LIMIT 5;"

try:
    # 2. Mở lại kết nối
    conn = mysql.connector.connect(
        host=host,
        database=database,
        user=username,
        password=password
    )
    
    # 3. Đưa dữ liệu thẳng vào Pandas DataFrame
    print("Đang truy vấn dữ liệu...")
    df = pd.read_sql(query, conn)
    
    print("✅ Đã tải dữ liệu thành công! Dưới đây là bảng xem trước:")
    # Hàm display() của Jupyter giúp hiển thị bảng dữ liệu rất đẹp và dễ nhìn
    display(df)
    
    # 4. Đóng kết nối
    conn.close()

except Exception as e:
    print(f"❌ Có lỗi khi truy vấn: {e}")
```
